```python
li={'content':[],'chungu_bi':[],'pi_bi':[]}
```

```python
def next_page(page):
    page_str = str(page).zfill(6)
    url = f'https://accident.knia.or.kr/example-content?index=2007-{page_str}'
    response = requests.get(url, verify=False)
    html_text = response.text
    html = bs(html_text, 'html.parser')
    
    # div가 없는 경우 다음 페이지로 이동
    if not html.find_all('div', class_='ex_view5_exp'):
        return
    
    bi = html.find_all('div', class_='ex_view5_exp')
    for i, b in enumerate(bi):
        if i % 2 == 0:
            without = b.text.replace('%', '')
            li['chungu_bi'].append(without)
        else:
            without = b.text.replace('%', '')
            li['pi_bi'].append(without)
    
    content = html.find_all('div', class_='ex_view2_exp')
    sentence = ""  
    for p in content:
        p_tag = p.find_all('p')
        for pp in p_tag:
            if not pp.find('strong'):
                sentence += pp.text + " " 
    li['content'].append(sentence.strip())
```

```python
for i in range(230,1000):
    next_page(i)
```
